```python
#Temel kütüphaneler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
planets = sns.load_dataset("planets") 
planets.head() 
```

```python
plt.figure(figsize=(23,6))
planets["method"].value_counts().plot.barh()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(data = planets)
```

```python
sns.catplot(data=planets)
```

```python
plt.figure(figsize=(15,6))
sns.distplot(planets.orbital_period)
```

```python
sns.catplot(kind="point",data =planets)
```

```python
sns.boxplot(data = planets)
```

```python
plt.figure(figsize=(15,6))
sns.scatterplot(data = planets)
```
